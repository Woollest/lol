```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a_val = 1234
b_val = 5678
```

```python
a = [int(d) for d in str(a_val)[::-1]]
b = [int(d) for d in str(b_val)[::-1]]
```

```python
print(a)
print(b)
```

```python
plt.stem(range(len(a)), a, linefmt = 'b-', label = 'a_digits')
plt.stem(range(len(b)), b, linefmt = 'r-', label = 'b_digits')
plt.title("Step 1: Digits as Discrete Signals")
plt.show()
```

```python
N = 16
A = np.fft.fft(a, N)
B = np.fft.fft(b, N)
```

```python
print(A)
print(B)
```

```python
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.plot(np.abs(A), 'bo-')
plt.title("FFT Spectrum of 'a'")
plt.subplot(1, 2, 2)
plt.plot(np.abs(B), 'ro-')
plt.title("FFT Spectrum of 'b'")
plt.show()
```

```python
C = A * B
c_res = np.fft.ifft(C).real.round().astype(int)
```

```python
print("繰り上がり前の各桁:", c_res[:len(a) + len(b) - 1])
```

```python
final_res = 0
for i, val in enumerate(c_res):
    final_res += val * (10**i)
```

```python
print("最終計算結果:", final_res)
print("正解確認:", a_val * b_val)
```
